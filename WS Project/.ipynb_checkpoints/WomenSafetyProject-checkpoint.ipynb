```python
import pandas as pd
import numpy as np
import random 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data.csv')
df.head(2)
```

```python
ThresholdT = 95
ThresholdP = 101
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Time',y='Temp',data=df)
plt.axhline(ThresholdT,color='r', linewidth=2)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Time',y='Pulse',data=df)
plt.axhline(ThresholdP,color='r', linewidth=2)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Time',y='Touch',data=df)
# plt.axhline(ThresholdT,color='r', linewidth=2)
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='Time',y = 'Pulse',hue = 'Temp', data = df,palette = 'bright')
```

```python
sns.regplot(x='Time',y = 'Temp', data = df)
```

```python
plt.figure(figsize=(10,5))
df.plot(x="Time", y=["Temp", "Pulse","Touch"])
```

```python
Temp = pd.read_csv('Temp.csv')
Pulse = pd.read_csv('Pulse.csv')
Touch = pd.read_csv('Touch.csv')
```

```python
print(Temp.head(2))
print(Pulse.head(2))
print(Touch.head(2))
```

```python
plt.figure(figsize=(20,15))
```

```python
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
plt.axhline(ThresholdP,color='r', linewidth=2)
sns.lineplot(x='Time',y='Pulse',data=Pulse)
plt.subplot(2,2,2)
plt.axhline(ThresholdT,color='r', linewidth=2)
sns.lineplot(x='Time',y='Temp',data=Pulse)
plt.subplot(2,2,3)
plt.axhline(1,color='r', linewidth=2)
sns.lineplot(x='Time',y='Touch',data=Pulse)
```

```python
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
plt.axhline(ThresholdP,color='r', linewidth=2)
sns.lineplot(x='Time',y='Pulse',data=Temp)
plt.subplot(2,2,2)
plt.axhline(ThresholdT,color='r', linewidth=2)
sns.lineplot(x='Time',y='Temp',data=Temp)
plt.subplot(2,2,3)
plt.axhline(1,color='r', linewidth=2)
sns.lineplot(x='Time',y='Touch',data=Temp)
```

```python
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
plt.axhline(ThresholdP,color='r', linewidth=2)
sns.lineplot(x='Time',y='Pulse',data=Touch)
plt.subplot(2,2,2)
plt.axhline(ThresholdT,color='r', linewidth=2)
sns.lineplot(x='Time',y='Temp',data=Touch)
plt.subplot(2,2,3)
plt.axhline(0.5,color='r', linewidth=2)
sns.lineplot(x='Time',y='Touch',data=Touch)
```

```python
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
plt.axhline(ThresholdP,color='r', linewidth=2)
sns.lineplot(x='Time',y='Pulse',data=df)
plt.subplot(2,2,2)
plt.axhline(ThresholdT,color='r', linewidth=2)
sns.lineplot(x='Time',y='Temp',data=df)
plt.subplot(2,2,3)
plt.axhline(0.5,color='r', linewidth=2)
sns.lineplot(x='Time',y='Touch',data=df)
```
